```python
#importar biblioteca:
import pandas as pd
import prettytable
import sqlite3

prettytable.DEFAULT = 'DEFAULT'
```

```python
#LEITURA DO ARQUIVO USANDO PANDAS:
df1 = pd.read_csv(r"/content/cousera_sql/ChicagoCrimeData.csv")
df2 = pd.read_csv(r"/content/cousera_sql/ChicagoPublicSchools.csv")
df3 = pd.read_csv(r"/content/cousera_sql/ChicagoCensusData.csv")
```

```python
conexao = sqlite3.connect('FinalDB.db')
```

```python
#TRANSFORMA O DATAFRAME EM UMA TABELA
df1.to_sql('ChicagoCrimeData', conexao, if_exists='replace', index=False)
df2.to_sql('ChicagoPublicSchools', conexao, if_exists='replace', index=False)
df3.to_sql('ChicagoCensusData', conexao, if_exists='replace', index=False)
```

```python
conexao.close()
```

```python
%%sql
SELECT
    "Elementary, Middle, or High School",
    round(AVG(SAFETY_SCORE), 1)
FROM ChicagoPublicSchools
GROUP BY "Elementary, Middle, or High School";
```

```python
%%sql select community_area_name, max(PERCENT_HOUSEHOLDS_BELOW_POVERTY) as max_poverty from ChicagoCensusData group by community_area_name order by max_poverty desc limit 5;
```

```python
%%sql select community_area_number
from ChicagoCensusData
ORDER by PERCENT_HOUSEHOLDS_BELOW_POVERTY DESC LIMIT 1;
```

```python
%%sql
SELECT community_area_number,
       COUNT(*) AS total_crimes
FROM ChicagoCrimeData
GROUP BY community_area_number
ORDER BY total_crimes DESC
LIMIT 1;
```

```python
%%sql
SELECT community_area_number
       COUNT(*) AS total_crimes
FROM ChicagoCrimeData
GROUP BY community_area_number
ORDER BY total_crimes DESC
LIMIT 1;
```
